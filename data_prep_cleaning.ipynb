```python
# this script cleans the original dataset
# data preparation > data cleaning
import pandas as pd
df = pd.read_excel("CAN_short_20220504.xlsx")
print(df.head())
```

```python
df.head(3)
```

```python
df = df.dropna(axis=1, how='all')
for i in range(1, 82):
    try:
        df = df.drop([f'Unnamed: {i}'], axis=1)
    except KeyError: continue

df.head()
```

```python
df.rename(columns={"Unnamed: 0": "Time"}, inplace=True)
df.head()
```

```python
df.head()
```

```python
#(df["Time"] - df["Time"][0]).total_seconds()
ref = df["Time"][0]
df["ReferenceTime"] = df.apply(lambda x: (x["Time"] - ref).total_seconds(), axis=1)
```

```python
df.head()
```

```python
df = df.iloc[0:969]
```

```python
col = df.pop('ReferenceTime')
df.insert(1, col.name, col)
df.head()
```

```python
df.to_excel("CAN_short_20220504_cleaned.xlsx", index=False)
```

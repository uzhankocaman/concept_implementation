```python
import pandas as pd
import time

i = 0

def access_data(sensor_data_generator):
        """
        Accesses data from the sensor data generator.
        """
        try:
            data = pd.DataFrame(next(sensor_data_generator), index=[0])
            return data
        except StopIteration:
            return None

def simulated_sensor(file_path, transmission_interval=0):
    data = pd.read_excel(file_path)
    data_records = data.to_dict('records')
    for record in data_records:
        yield record
        time.sleep(transmission_interval)

file_path = "../data/CAN_short_20220504.xlsx"
sensor_data_generator = simulated_sensor(file_path, 0)

while True:
    i=i+1
    data = access_data(sensor_data_generator)
    print(data["Unnamed: 0"])
print(i)
```

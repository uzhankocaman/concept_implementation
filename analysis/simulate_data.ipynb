```python
import pandas as pd
import time

def simulated_sensor(file_path):
    data = pd.read_excel(file_path)
    data_records = data.to_dict('records')
    for record in data_records:
        yield record
        time.sleep(1)
```

```python
for sensor_data in simulated_sensor("CAN_short_20220504_cleaned.xlsx"):
    print(sensor_data)
```

```python
def access_data(simulated_data_generator, store_func):
    for data_row in simulated_data_generator:
        store_func(data_row)

def store_data(dataframe, data_row):
    return dataframe.append(data_row, ignore_index=True)
```
